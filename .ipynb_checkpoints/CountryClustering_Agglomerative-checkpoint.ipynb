## Single Linkage Clustering
This part will focus on *Agglomerative clustering* using **Single Linkage Criterion**.

Single Linkage Critereon: is agglomerative clustering method where we merge the pair of clusters (s) and (r) that has the smallest distance. And the new distance between the new formed cluster (rs) and a point i is $d[(rs),i] = min(d[(r),i],d[(s),i])$.

### Clustering using the  *AgglomerativeClustering_fs* class
In this section, we will use the from scratch class *AgglomerativeClustering_fs*. In the implementation we got inspired from the Scipy documentation and followed the same process.
So for clustering we will follow this steps: 
1. Create an instance of the classe, for Single Linkage we will set the linkage attribute with the value 'single'.
2. Fit the model to the dataset, the fit method computes the linkage matrix and set the attribute L.
3. Get the labels using the method get_labels precising the maximal distance value between two clusters to merge them.

#### Computing the Linkage matrix
***Linkage matrix*** is well described in the Scipy documentation [[1]](#ref1). It's a matrix counting n-1 rows and four columns, where n is the number of original observations. At a row * I*, the first two columns refer to the two clusters that merged to form the cluster n+i, the third column to the distance between clusters, and the last column to the number of elements in the new cluster; the sum of the elements of the two clusters. In our implementation we return the same matrix.

```python
# Clustering using the AgglomerativeClustering_fs
model = AgglomerativeClustering_fs.AgglomerativeClustering_fs(linkage='single')
model.fit(X)
L = model.L
print('The first 6 rows of our linkage matrix:')
print(L[:6])
print('The last 6 rows of our linkage matrix:')
print(L[-6:])
```

The distances in the resultant Linkage matrix ranges from 0.27 to 6.84.<br>
The higher the distance between formed clusters, the higher the disparities between the countries of the group.
At the head of our Linkage matrix, we note that if we choose a maximal distance of 0.5, we will get one cluster of 4 elements, and the rest of the countries will be singletons. For our clusterization to draw some meaningful conclusions, we will need to choose a maximal distance that reduces the number of singletons.<br>
In the tail of our Linkage matrix, We can see at 3.7 that we merged two singletons. It implies that if we choose a 3.8 as the maximal distance, we will obtain: one cluster of 163 countries, another with two countries, and two singletons. We need to highlight two remarks:
- In our dataset, even with a high maximal distance, we will get singletons. The last line is a marge between a 166 elements cluster and a singleton.
- The cluster is highly populated, that we can't draw any meaningful patterns from it.

#### Choosing the Maximal Distance to Merge clusters

To determine the optimal maximal distance value. We will use the metrics **Davies Bouldin Index** and **Silhouette Coefficient**.
- **Davies Bouldin Index**: This metric evaluates the ratio between the within-cluster distance and that of intra-clusters. We are looking for the maximal distance value that lowers the score: it translates a low dissimilarity between the elements of the same group and a high distance between clusters.
- **Silhouette Score**: This metric computes the mean of the Silhouette index of the samples, where a sample is the observations of a cluster. We define the Silhouette Coefficient for a sample $C_i$ as the ratio difference between a: The within-cluster distance, and b: the distance to the nearest cluster to $C_i$. And the max of this to values. $\frac{(a-b)}{max(a,b)}$. This score ranges between -1 and 1, the best score would be 1 and worst -1.

```python
#Choosing the max distance that optimise the davies bouldin score
max_distances = np.arange(0.75,int(max(L[:,2]))+1,0.25)   #list(range(1,int(max(Z_single[:,2]))+1)) #
davies_bouldin_scores = []
silhouette_scores = []
nbr_clusters = []

for dist in max_distances:
        labels_test = model.get_labels(max_dist = dist)
        nbr_clusters.append(max(labels_test))
        davies_bouldin_scores.append(davies_bouldin_score(X, labels_test))
        silhouette_scores.append(silhouette_score(X, labels_test))
        
minIdx_DBI = davies_bouldin_scores.index(min(davies_bouldin_scores))
maxIdx_SI = silhouette_scores.index(max(silhouette_scores))
            
print('The minimum Davies Bouldin Score is {} with a maxdistance {}'.format(min(davies_bouldin_scores),max_distances[minIdx_DBI]))
print('The maximum Silhouette Score is {} with a maxdistance {}'.format(max(silhouette_scores),max_distances[maxIdx_SI]))


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6),squeeze=False)



axes[0][0].plot(max_distances,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][0].plot(max_distances,silhouette_scores,label = 'Silhouette Score')
axes[0][0].legend()
axes[0][0].set_title('Scores per Max Distance for the Linkage method Single')

axes[0][1].plot(nbr_clusters,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][1].plot(nbr_clusters,silhouette_scores,label = 'Silhouette Score')
axes[0][1].legend()
axes[0][1].set_title('Scores per Number of Clusters for the Linkage method Single')



```

Bought the Davies Bouldin Score and Silhouette Score reached their optimal values at the same maximal distance 4.25. From the tail of our linkage matrix, we can see that at this maximal distance value, we will obtain two clusters:<br>
- First cluster: One cluster of 166 countries. 
- Second cluster: One singleton country.

Let's visualize our clustering tree using the dendrogram and compare the suggested maximal distance value and other potential values.

```python
fig = plt.figure(figsize=(20, 10))
dn = dendrogram(L,truncate_mode='level',p=15,get_leaves=False,leaf_font_size=12.,)
plt.axhline(y=2.5,c='r')
plt.axhline(y=3.5,c='r')
plt.axhline(y=4.25,c='r')
plt.show()
```

To have a readable graph, we have chosen to truncate our dendrogram at level 15, so we don't have all the formation steps of the cluster. From the Davies Bouldin Score and Silhouette Score graph, we can see the two scores intersect at 2.5, so we have chosen to compare it to the suggested maximal distance of 4.25, and we added a second potential value of 3.5.

- Maximal distance 2.5: We can see six singleton countries that didn't merge with any cluster, and we have a group with 161 countries.
- Maximal distance 3.5: At this level, we can see two other countries joining the general cluster, and the other four countries remain singletons.
- Maximal distance 4.25: The dendrogram confirms our linkage matrix read. We notice that all the countries merged in one general cluster, except one country.

We can notice that with the three potential maximal distance values, the single-linkage clusterization returns one cluster and few singleton countries. The threshold value 4.25, which optimizes the Davies Bouldin and Silhouette scores, returns one cluster grouping all the countries except one. We can see in the graph that the distance between the general cluster in green and the singleton country in blue is high. It indicates that the similarities between this country and the others are minimal. But other than highlighting this country, we will not have any meaningful groups.

If we choose 2.5 or 3.5 as the maximal distance value, we get a general cluster and few singletons. We will choose the one that minimizes the number of singletons which is 3.5.

#### Get the labels

```python
max_dist = 3.5
labels = model.get_labels(max_dist=max_dist)
country_data['Single_Clusters_1'] = labels
print('The labels of the Single-Linkage Clusterization : {}'.format(np.unique(labels)))
country_data.head()
```

```python
palette = sns.color_palette(None, max(labels))
plt.figure()

for i, color in enumerate(palette):
    j = np.where(labels == i+1)
    plt.scatter(X_pca[j, 0], X_pca[j, 1], color=color, alpha=.8)

plt.show()
```

We can see the 163 countries belonging to the first cluster in blue and the singleton countries in orange, green, red, and violet. This clusterization highlighted the most distant datapoints (countries), similarly as if we are detecting outliers. But was not able to partition the other countries.     
### Clustering using Scipy and Sklearn
In this section, we will use the Scipy and Sklearn implementations for building our single-linkage agglomerative clusters. And we will compare the results of each library to our own.

#### Scipy
We will use the Scipy *linkage function*, which takes as a parameter an array of Pairwise distances between observations and returns a linkage matrix. Then we retrieve the labels of our countries through the function *fcluster* with the parameters:
- The Linkage matrix that the *linkage function* returned
- The criterion distance to base the merge of our clusters on the distance between them. 
- The maximal distance with the value we used earlier, which is equal to 4.25. 

```python
# Clustering using the scipy linkage
Z = linkage(pdist(X), 'single')
if (np.round(Z,12) == np.round(L,12)).all():
    print('The From Scratch Linkage Matrix is equal to The Scipy Linkage Matrix')
else:
    print('The From Scratch Linkage Matrix is not equal to The Scipy Linkage Matrix')

labels = fcluster(Z, t=max_dist, criterion='distance')
country_data['Single_Clusters_2'] = labels

for i in np.unique(country_data.Single_Clusters_1):
    if len(np.unique(country_data[country_data['Single_Clusters_1']==i].Single_Clusters_2)) >1 :
        print('The From Scrtach clusters and the Scipy clusters are different.')
        break
    else:
        continue 

if i == max(country_data.Single_Clusters_1):
     print('The From Scrtach clusters and the Scipy clusters are the same.')

country_data.head()
```

#### Sklearn
To build the sklearn clusters, we will use the class *AgglomerativeClustering*. We will start by creating an instance of this class with the parameters:
- linkage: We use this parameter to specify the linkage method. In our case, we will use the 'single' criterion.
- distance_threshold: The maximal distance value to merge clusters. We will use the same value as AgglomerativeClustering_fs and Scipy.
- n_clusters: Number of clusters. Because we have determined the distance threshold, we will not have to specify this parameter.  We will set it to None.

```python
model_sk = AgglomerativeClustering(distance_threshold=max_dist, n_clusters=None, linkage= 'single')
country_data['Single_Clusters_3'] = model_sk.fit_predict(X)

for i in np.unique(country_data.Single_Clusters_1):
    if len(np.unique(country_data[country_data['Single_Clusters_1']==i].Single_Clusters_3)) >1 :
        print('The From Scrtach clusters and the Sklearn clusters are different.')
        break
    else:
        continue 

if i == max(country_data.Single_Clusters_1):
     print('The From Scrtach clusters and the Sklearn clusters are the same.')

country_data.head()
```

Our implementation yield the same results as the most popular libraries.<br>
In the results section, we will compare the different linkage methods to find the best suited for our study case.

## Complete Linkage Clustering
This part will focus on *Agglomerative clustering* using **Complete Linkage Criterion**.

Complete Linkage Criterion: is an agglomerative clustering method where we merge the pair of clusters (s) and (r) with the smallest distance. And the new distance between the new formed cluster (rs) and a point i is $d[(rs),i] = max(d[(r),i],d[(s),i])$. The difference between single and complete linkage is that the first one considers the distance between two clusters as the distance between their closest points, and the latter the farthest points.

### Clustering using Cluserting_Algorithms module
In this section, we will test our class AgglomerativeClustering_fs. We will follow the same steps, but we will create our model instance with the attribute linkage='compelete'.<br>

#### Computing the Linkage matrix

```python
# Clustering using the AgglomerativeClustering_fs
model = AgglomerativeClustering_fs.AgglomerativeClustering_fs(linkage='complete')
model.fit(X)
L = model.L
print('The first 6 rows of our linkage matrix:')
print(L[:6])
print('The last 6 rows of our linkage matrix:')
print(L[-6:])
```

When comparing the single and complete linkage matrices, We can notice that the distance values in the latter range are between 0.27 and 14.98, whereas in the first, they range between 0.27 and 6.84.<br>
We can also note that the last two clusters to be joined had 55 and 112 countries each, which indicates that the clusters are more populated, and thereby our conclusions will be more meaningful than those we draw from the single linkage.<br>

#### Choosing the Maximal Distance to Merge clusters

```python
#Choosing the max distance that optimise the davies bouldin score
max_distances = np.arange(0.75,int(max(L[:,2]))+1,0.25)   #list(range(1,int(max(Z_single[:,2]))+1)) #
davies_bouldin_scores = []
silhouette_scores = []
nbr_clusters = []

for dist in max_distances:
        labels_test = model.get_labels(max_dist = dist)
        nbr_clusters.append(max(labels_test))
        davies_bouldin_scores.append(davies_bouldin_score(X, labels_test))
        silhouette_scores.append(silhouette_score(X, labels_test))
        
minIdx_DBI = davies_bouldin_scores.index(min(davies_bouldin_scores))
maxIdx_SI = silhouette_scores.index(max(silhouette_scores))
            
print('The minimum Davies Bouldin Score is {} with a maxdistance {}'.format(min(davies_bouldin_scores),max_distances[minIdx_DBI]))
print('The maximum Silhouette Score is {} with a maxdistance {}'.format(max(silhouette_scores),max_distances[maxIdx_SI]))



fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6),squeeze=False)


axes[0][0].plot(max_distances,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][0].plot(max_distances,silhouette_scores,label = 'Silhouette Score')
axes[0][0].legend()
axes[0][0].set_title('Scores per Max Distance for the Linkage method Complete')

axes[0][1].plot(nbr_clusters,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][1].plot(nbr_clusters,silhouette_scores,label = 'Silhouette Score')
axes[0][1].legend()
axes[0][1].set_title('Scores per Number of Clusters for the Linkage method Complete')
plt.show()
```

The Davies Bouldin Score and Silhouette Score reached their optimal values at different distance thresholds: 
- The Davies Bouldin Score: with this metric lower values is better. We can see in the graph that the Davies Bouldin Scores increase with the distance. So this coefficient reaches its optimal values with small distance thresholds.
- The Silhouette Score: these metric values range between -1 and 1, the best being 1. The Silhouette Scores in the graph seem slightly increasing with the distance.
Referring to the number of clusters, choosing the DBI optimal point doesn't seem reasonable since it will return 140 groups. To have more visibility, We will draw our dendrogram with the two suggested maximal distances and another potential threshold of 8.5.

```python
fig = plt.figure(figsize=(20, 10))
dn = dendrogram(L,truncate_mode='level',p=15,get_leaves=False,leaf_font_size=12.,)
plt.axhline(y=0.75,c='r')
plt.axhline(y=8.5,c='r')
plt.axhline(y=11,c='r')
plt.show()
```

## UPGMA Linkage Clustering
This part will focus on *Agglomerative clustering* using **UPGMA Linkage Criterion**.

UPGMA Linkage Critereon: or **U**nweighted **P**air **G**roup **M**ethod with **A**rithmetic mean, is agglomerative clustering method where we merge the pair of clusters (s) and (r) that has the smallest distance. And the new distance between the new formed cluster (rs) and a cluster (k) is ${\displaystyle {\frac {1}{|(rs)|\cdot |(k)|}}\sum _{i\in (rs)}\sum _{j\in (k)}d(i,j).}$

### Clustering using Cluserting_Algorithms module
In this section, we will test our class AgglomerativeClustering_fs. We will follow the same steps, but we will create our model instance with the attribute linkage='UPGMA'.<br>

#### Computing the Linkage matrix

```python
# Clustering using the AgglomerativeClustering_fs
model = AgglomerativeClustering_fs.AgglomerativeClustering_fs(linkage='UPGMA')
model.fit(X)
L = model.L
print('The first 6 rows of our linkage matrix:')
print(L[:6])
print('The last 6 rows of our linkage matrix:')
print(L[-6:])
```

#### Choosing the Maximal Distance to Merge clusters

```python
#Choosing the max distance that optimise the davies bouldin score
max_distances = np.arange(0.75,int(max(L[:,2])),0.25)   #list(range(1,int(max(Z_single[:,2]))+1)) #
davies_bouldin_scores = []
silhouette_scores = []
nbr_clusters = []

for dist in max_distances:
        labels_test = model.get_labels(max_dist = dist)
        nbr_clusters.append(max(labels_test))
        davies_bouldin_scores.append(davies_bouldin_score(X, labels_test))
        silhouette_scores.append(silhouette_score(X, labels_test))
        
minIdx_DBI = davies_bouldin_scores.index(min(davies_bouldin_scores))
maxIdx_SI = silhouette_scores.index(max(silhouette_scores))
            
print('The minimum Davies Bouldin Score is {} with a maxdistance {}'.format(min(davies_bouldin_scores),max_distances[minIdx_DBI]))
print('The maximum Silhouette Score is {} with a maxdistance {}'.format(max(silhouette_scores),max_distances[maxIdx_SI]))



fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6),squeeze=False)


axes[0][0].plot(max_distances,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][0].plot(max_distances,silhouette_scores,label = 'Silhouette Score')
axes[0][0].legend()
axes[0][0].set_title('Scores per Max Distance for the Linkage method UPGMA')

axes[0][1].plot(nbr_clusters,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][1].plot(nbr_clusters,silhouette_scores,label = 'Silhouette Score')
axes[0][1].legend()
axes[0][1].set_title('Scores per Number of Clusters for the Linkage method UPGMA')
plt.show()
```

The two Indices reached their optimal values at the same distance threshold which is equal to 8.75.

```python
fig = plt.figure(figsize=(20, 10))
dn = dendrogram(L,truncate_mode='level',p=12,get_leaves=False,leaf_font_size=12.,)
plt.axhline(y=8.75,c='r')
plt.axhline(y=6.75,c='r')
plt.show()
```

- Maximal distance 6.75: We can see two singletons, one cluster with three countries, and all the remaining countries are in one group.
- Maximal distance 8.75: We obtained one singleton country and all the other countries are grouped in one cluster.
#### Get the labels

We will opt for the threshold value of 6.75.

```python
max_dist = 6.75
labels = model.get_labels(max_dist=max_dist)

print('The labels of the UPGMA-Linkage Clusterization : {}'.format(np.unique(labels)))
country_data['UPGMA_Clusters_1'] = labels
columns = list(features)+['UPGMA_Clusters_1']
country_data[columns].head()
```

```python
palette = sns.color_palette(None, max(labels))
plt.figure()

for i, color in enumerate(palette):
    j = np.where(labels == i+1)
    plt.scatter(X_pca[j, 0], X_pca[j, 1], color=color, alpha=.8)

plt.show()
```

## WPGMA Linkage Clustering
This part will focus on *Agglomerative clustering* using **WPGMA Linkage Criterion**.

WPGMA Linkage Critereon: or **W**eighted **P**air **G**roup **M**ethod with **A**rithmetic mean, is agglomerative clustering method where we merge the pair of clusters (s) and (r) that has the smallest distance. And the new distance between the new formed cluster (rs) and a cluster (k) is ${\displaystyle {\frac {d(r,k)+d(s,k)}{2}}.}$

### Clustering using Cluserting_Algorithms module
In this section, we will test our class AgglomerativeClustering_fs. We will follow the same steps, but we will create our model instance with the attribute linkage='WPGMA'.<br>

#### Computing the Linkage matrix

```python
# Clustering using the AgglomerativeClustering_fs
model = AgglomerativeClustering_fs.AgglomerativeClustering_fs(linkage='WPGMA')
model.fit(X)
L = model.L
print('The first 6 rows of our linkage matrix:')
print(L[:6])
print('The last 6 rows of our linkage matrix:')
print(L[-6:])
```

#### Choosing the Maximal Distance to Merge clusters

```python
#Choosing the max distance that optimise the davies bouldin score
max_distances = np.arange(0.75,int(max(L[:,2])),0.25)   #list(range(1,int(max(Z_single[:,2]))+1)) #
davies_bouldin_scores = []
silhouette_scores = []
nbr_clusters = []

for dist in max_distances:
        labels_test = model.get_labels(max_dist = dist)
        nbr_clusters.append(max(labels_test))
        davies_bouldin_scores.append(davies_bouldin_score(X, labels_test))
        silhouette_scores.append(silhouette_score(X, labels_test))
        
minIdx_DBI = davies_bouldin_scores.index(min(davies_bouldin_scores))
maxIdx_SI = silhouette_scores.index(max(silhouette_scores))
            
print('The minimum Davies Bouldin Score is {} with a maxdistance {}'.format(min(davies_bouldin_scores),max_distances[minIdx_DBI]))
print('The maximum Silhouette Score is {} with a maxdistance {}'.format(max(silhouette_scores),max_distances[maxIdx_SI]))



fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6),squeeze=False)


axes[0][0].plot(max_distances,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][0].plot(max_distances,silhouette_scores,label = 'Silhouette Score')
axes[0][0].legend()
axes[0][0].set_title('Scores per Max Distance for the Linkage method WPGMA')

axes[0][1].plot(nbr_clusters,davies_bouldin_scores,label = 'Davies Bouldin Score')
axes[0][1].plot(nbr_clusters,silhouette_scores,label = 'Silhouette Score')
axes[0][1].legend()
axes[0][1].set_title('Scores per Number of Clusters for the Linkage method WPGMA')
plt.show()
```

In the interval between 8.25 and 10.25, we can see that bought indices are close to their respective optimal values. In addition to the suggested thresholds of 10.5 and 0.75, we will visualize the threshold line of 8.25.

```python
fig = plt.figure(figsize=(20, 10))
dn = dendrogram(L,truncate_mode='level',p=12,get_leaves=False,leaf_font_size=12.,)
plt.axhline(y=10.5,c='r')
plt.axhline(y=8.25,c='r')
plt.axhline(y=0.75,c='r')
plt.show()
```

- Maximal distance 0.75: We can see a majority of singletons and pairs.
- Maximal distance 8.25: We can see four clusters including two singletons.
- Maximal distance 10.5: We can notice two clusters. One counts two countries, and the other groups the remaining countries.

Centroid Linkage Critereon: is agglomerative clustering method where we merge the pair of clusters (s) and (r) that has the smallest distance. The new distance is the distance between the centroid of the new formed cluster and the centroids of the existing clusters.

Ward Linkage Critereon: is an agglomerative clustering method where at iteration i the most similar pair of clusters (r,s) is the pair that minimizes the within-cluster variance after merge.

## Results
In this section, we will compare the different method results. To do so, we will compute the **Calinski Harabasz Score**.

- **Calinski Harabasz Score** is a metric that evaluates the dispersion of the clusters. It is the ratio between the inter-Clusters dispersion and the intra-Cluster dispersion. Higher values of this metric imply a well-defined partition.

```python
metrics = []
methods = ['single','complete','UPGMA','WPGMA','centroid','ward']
labels_permethods = ['Single_Clusters_1','Complete_Clusters_1','UPGMA_Clusters_1','WPGMA_Clusters_1','Centroid_Clusters_1','Ward_Clusters_1']
for method,labels in zip(methods,labels_permethods):
    metrics.append([method, 
                    calinski_harabasz_score(X, country_data[labels])
                   ])
metrics = pd.DataFrame(metrics,
            columns=['Method','Calinski Harabasz Score'])
metrics.sort_values(by = 'Calinski Harabasz Score',
                    ascending = False)
```

We will base our clusters on the **ward** criterion since it maximizes the Calinski Harabasz Score.

```python
clusters_labels = np.unique(country_data['Ward_Clusters_1'])

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(28, 25))

country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['exports'],by='Ward_Clusters_1', fontsize=12,ax=axes[0][0])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['imports'],by='Ward_Clusters_1',ax=axes[0][1])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['health'],by='Ward_Clusters_1',ax=axes[0][2])

country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['income'],by='Ward_Clusters_1',ax=axes[1][0])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['inflation'],by='Ward_Clusters_1',ax=axes[1][1])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['gdpp'],by='Ward_Clusters_1',ax=axes[1][2])

country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['child_mort'],by='Ward_Clusters_1',ax=axes[2][0])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['life_expec'],by='Ward_Clusters_1',ax=axes[2][1])
country_data[country_data['Ward_Clusters_1'].isin(clusters_labels)].boxplot(column=['total_fer'],by='Ward_Clusters_1',ax=axes[2][2])

fig.suptitle('Socio-Economic and Health Factors', fontsize=16)
plt.savefig('./Img/Socio_Economic_Health_Factors.png')
```

- Cluster 1: This cluster shows some signs of economic unbalance. We see that 75% of these countries' imports exceed 50% of their respective GDP per Capita, whereas, in three-quarters of this group, the exports represent 60% or less of their GDP per capita. Their GDP and Income are also one of the lowest 75% of these countries have 20000 or less as a net income per person. If we look into the health factors, we can see that they are moderate: The population of 75% of the elements of this group has a life expectancy between 70 and 80 years. Few outliers child mortality can reach 100 deaths per 1000 live births, and 75% of these countries have 25 or fewer deaths per 1000 live births. 
- Cluster 2: This group has one of the highest Income, GDP, Life expectancy, and one of the lowest child mortality rates.
- Cluster 3: This group has the lowest Income, GDP, Life expectancy, and the highest children mortality.
- Cluster 4: This group is close to cluster 3, we can see that the income and GDP are one of the lowest, and childhood mortality is second worse.
- Cluster 5: This cluster groups rich countries with high incomes and GDP. It also scores the lowest children mortality and highest life expectancy.
